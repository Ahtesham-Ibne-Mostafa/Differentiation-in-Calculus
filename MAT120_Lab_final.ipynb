```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Define the function
x = sp.Symbol('x')
f = 2*x**4 + x**3 + 4*x

# Calculate the first derivative
f_prime = sp.diff(f, x)

print('f_prime: ',f_prime)

# Calculate the second derivative
f_prime_prime = sp.diff(f_prime, x)

print ('f_prime_prime: ',f_prime_prime)

# Calculate the third derivative
f_prime_prime_prime = sp.diff(f_prime_prime, x)

print('f_prime_prime_prime: ',f_prime_prime_prime)

# Find the value of the first derivative at x = 2
x_value = 4
f_prime_value = f_prime.subs(x, x_value)
print("The value of the first derivative at x = {} is {}".format(x_value, f_prime_value))

# Plot the graph of the first derivative on the interval (-5, 6)
x_range = np.linspace(-5, 6, 100)
f_prime_range = sp.lambdify(x, f_prime, 'numpy')
y_range = f_prime_range(x_range)

plt.plot(x_range, y_range)
plt.xlabel("x")
plt.ylabel("dy/dx")
plt.title("Graph of the First Derivative of y = 2x^4 + x^3 + 4x")
plt.show()
```
