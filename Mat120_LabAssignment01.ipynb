```python
from numpy import*
from sympy import*
```

```python
x = symbols('x')
```

```python
# Function 1
f1 = 9*x**3 - 7*x**2 + 3*x + 10
df1 = diff(f1,x)
critical_points_f1 = solve(df1, x)

print("Function 1: y =", f1)
plot(f1, (x,-5,6))

# Check second derivative


d2f1 = diff(f1, x,2)
if len(critical_points_f1) == 0 :
  print("NO maxima or minima")
else:
  print("Critical points:", critical_points_f1)
  cnt = 0
  for x_val in critical_points_f1:
    cnt += 1
    if str(type(x_val)) == "<class 'sympy.core.add.Add'>":
      print(f"No maxima or minima at {cnt}th critical point ")

    elif d2f1.subs(x, x_val) > 0:
      print("Local minima at x =", x_val,",y =", f1.subs({x:x_val}).evalf())
    elif d2f1.subs(x, x_val) < 0:
        print("Local maxima at x =", x_val,",y =", f1.subs({x:x_val}).evalf())
print()
```
